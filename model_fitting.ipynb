```
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```
def func(x, a, b):
  return a * -np.cos(x) + b
```

```
x = np.arange(0, 20, 0.02)
y = func(x, 10, 0)

plt.plot(x, y)
plt.show()
```

```
np.random.seed(1)
yn = y + 0.9 * np.random.normal(size=len(x))

plt.plot(x, yn)
plt.show()
```

```
popt, pcov = curve_fit(func, x, yn)
print(popt)
print(pcov)

plt.plot(x, y, 'g-', label='orig')
plt.plot(x, yn, label='noise')
plt.plot(x, func(x, *popt), 'r-', label='best fit', linewidth=2)
plt.legend(loc=2)
plt.show()
```
